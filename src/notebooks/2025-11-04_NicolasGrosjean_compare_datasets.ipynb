## Merge C2C and TDG with distance

```python
interesting_columns = [
    "navitia_id",
    "name",
    "network",
    "agency_name",
    "stop_id",
    "stop_name",
    "distance",
]
c2c_tdg_stops_gdf = gpd.sjoin_nearest(
    c2c_expanded_stops_gdf.to_crs(EPSG_WEB_MERCATOR),
    tdg_stops_gdf.to_crs(EPSG_WEB_MERCATOR),
    how="left",
    distance_col="distance",
)
c2c_tdg_stops_gdf[interesting_columns].head()
```

```python
# Remove () and put lower in C2C names
c2c_tdg_stops_gdf["name2"] = c2c_tdg_stops_gdf["name"].apply(
    lambda s: s[: s.index("(")].lower()
)
c2c_tdg_stops_gdf["name2"]
```

```python
# Put lower in TDG names
c2c_tdg_stops_gdf["stop_name2"] = c2c_tdg_stops_gdf["stop_name"].apply(lambda s: s.lower())
```

```python
# Add fuzzymatching on names
interesting_columns2 = [
    "navitia_id",
    "name",
    "network",
    "agency_name",
    "stop_id",
    "stop_name",
    "distance",
    "fuzz_ratio",
]
c2c_tdg_stops_gdf["fuzz_ratio"] = c2c_tdg_stops_gdf.apply(
    lambda row: fuzz.ratio(row["name2"], row["stop_name2"]), axis=1
)
c2c_tdg_stops_gdf[interesting_columns2].head()
```

```python
# Explore big distance to see where to cut
c2c_tdg_stops_gdf_sorted = c2c_tdg_stops_gdf[interesting_columns2].sort_values("distance")
c2c_tdg_stops_gdf_sorted[c2c_tdg_stops_gdf_sorted["distance"] > 50].head(10)
```

```python
c2c_tdg_max_dist = 200  # m
print(f"Remove matching at distance bigger than {c2c_tdg_max_dist}m")
```

```python
import matplotlib.pyplot as plt

c2c_tdg_stops_gdf2 = c2c_tdg_stops_gdf[c2c_tdg_stops_gdf["distance"] < c2c_tdg_max_dist]
plt.scatter(c2c_tdg_stops_gdf2["distance"], c2c_tdg_stops_gdf2["fuzz_ratio"])
plt.xlabel("distance")
plt.ylabel("fuzz_ratio on names")
plt.show()
```

```python
# Explore bad fuzzy matching
c2c_tdg_stops_gdf2_sorted = c2c_tdg_stops_gdf2[
    c2c_tdg_stops_gdf2["fuzz_ratio"] < 60
].sort_values("fuzz_ratio", ascending=False)
c2c_tdg_stops_gdf2_sorted[interesting_columns2]
```

```python
min_fuzz_ratio = 0
max_fuzz_ratio_dist = c2c_tdg_max_dist
print(
    f"Remove matching when distance is bigger than {max_fuzz_ratio_dist}m and fuzz ratio lower than {min_fuzz_ratio}"
)
```

```python
c2c_tdg_stops_gdf3 = c2c_tdg_stops_gdf2[
    (c2c_tdg_stops_gdf2["fuzz_ratio"] > min_fuzz_ratio)
    | (c2c_tdg_stops_gdf2["distance"] < max_fuzz_ratio_dist)
]
```

```python
merge_nb = len(c2c_tdg_stops_gdf3["navitia_id"].unique())
total_nb = len(c2c_expanded_stops_gdf["navitia_id"].unique())
print(f"{merge_nb} C2C stops merged on {total_nb} ({merge_nb / total_nb:.0%})")
```

# TDG data at less than Xkm of an activity

```python
distance_threshold_km = 3
```

```python
department_code = "38"
```

```python
department_activity_df = (
    gpd.sjoin(
        activity_gdf.to_crs(EPSG_WGS84),
        area_gdf[area_gdf["code"] == department_code],
    )
    .loc[:, ["Id wp"]]
    .drop_duplicates()
    .merge(activity_gdf, on="Id wp", how="inner")
)
department_activity_gdf = gpd.GeoDataFrame(department_activity_df, geometry="geometry").set_crs(
    EPSG_WGS84
)
print(
    f"Filter on department {department_code}: keep {len(department_activity_gdf)} on {len(activity_gdf)} activity points"
)
```

```python
distances_df = all_distances_df[all_distances_df["distance_m"] < distance_threshold_km * 1000]
nb_lines_before_drop = len(distances_df)
distances_df = distances_df[~pd.isnull(distances_df["distance_m"])]
distances_df = distances_df[
    distances_df["Id wp"].isin(department_activity_df["Id wp"].unique())
]
print(f"{len(distances_df):,} distances kept on {nb_lines_before_drop:,}".replace(",", " "))
distances_df.head()
```

```python
tdg_points = distances_df[~pd.isnull(distances_df["gtfs_id"])]["gtfs_id"].unique()
print(
    f"{len(tdg_points):,} transport data gouv bus stops are less than {distance_threshold_km}km of an activity on {tdg_stops_gdf['stop_id'].nunique():,} bus stops".replace(
        ",", " "
    )
)
```

```python
tdg_interesting_columns = ["stop_id", "stop_name", "stop_lines", "agency_name", "geometry"]
tdg_distances = distances_df.merge(
    tdg_stops_gdf[tdg_interesting_columns], left_on="gtfs_id", right_on="stop_id", how="inner"
)
tdg_mask = tdg_distances["distance_m"] == tdg_distances.groupby(["Id wp", "stop_lines"])[
    "distance_m"
].transform("min")
tdg_neareast_stops_by_line_gdf = tdg_stops_gdf[
    tdg_stops_gdf["stop_id"].isin(tdg_distances[tdg_mask]["stop_id"].unique())
]
print(
    f"{len(tdg_neareast_stops_by_line_gdf):,} tdg bus stops are the nearest activity points for their bus line (on {len(tdg_stops_gdf):,} bus stops)".replace(
        ",", " "
    )
)
```

```python
c2c_tdg_stops_gdf3["lines"] = c2c_tdg_stops_gdf3["stoparea_id_and_line"].apply(
    lambda l: [e["line"] for e in l]
)
```

```python
def list_to_str_with_br(l: list[str], line_number_limit: int = 10) -> str:
    list_to_process = l[:line_number_limit]
    if len(l) > line_number_limit:
        list_to_process += ["..."]
    return "<br>".join(list_to_process)


def get_tdg_line_names(lines: str) -> str:
    lines_json = "[" + lines.replace("'", '"') + "]"
    # TODO Remove this bugfix
    lines_json = lines_json.replace('Mobi"Aix', "Mobi'Aix")
    data = json.loads(lines_json)
    return list_to_str_with_br([d["line_name"] for d in data])
```

```python
bus_stop_nb = {
    "TDG": len(tdg_neareast_stops_by_line_gdf),
    "C2C_TDG": len(c2c_tdg_stops_gdf3["navitia_id"].unique()),
}
fields_by_source = {
    "TDG": ["stop_id", "stop_name", "agency_name", "stop_lines"],
    "C2C_TDG": [
        "stop_id",
        "stop_name",
        "agency_name",
        "stop_lines",
        "navitia_id",
        "name",
        "network",
        "lines",
    ],
}
aliases_by_source = {
    "TDG": [f"<b>Arrêt de bus TDG</b><br>ID :", "Nom :", "Réseau :", "Lignes :"],
    "C2C_TDG": [
        f"<b>Arrêt de bus TDG</b><br>ID :",
        "Nom :",
        "Réseau :",
        "Lignes :",
        f"<b>Arrêt de bus C2C</b><br>ID :",
        "Nom :",
        "Réseau :",
        "Lignes :",
    ],
}
color_by_source = {"TDG": "red", "C2C_TDG": "blue"}
tdg_stops_to_plot = tdg_neareast_stops_by_line_gdf[
    fields_by_source["TDG"] + ["geometry"]
].copy()
tdg_stops_to_plot["stop_lines"] = tdg_neareast_stops_by_line_gdf["stop_lines"].apply(
    get_tdg_line_names
)
c2c_tdg_stops_to_plot = c2c_tdg_stops_gdf3[fields_by_source["C2C_TDG"] + ["geometry"]].copy()
c2c_tdg_stops_to_plot["stop_lines"] = c2c_tdg_stops_to_plot["stop_lines"].apply(
    get_tdg_line_names
)
c2c_tdg_stops_to_plot["lines"] = c2c_tdg_stops_to_plot["lines"].apply(list_to_str_with_br)
valid_navitia_ids = gpd.sjoin(
    c2c_tdg_stops_to_plot.to_crs(EPSG_WGS84),
    area_gdf[area_gdf["code"] == department_code].to_crs(EPSG_WGS84),
)["navitia_id"].drop_duplicates()
c2c_tdg_stops_to_plot = c2c_tdg_stops_to_plot[
    c2c_tdg_stops_to_plot["navitia_id"].isin(valid_navitia_ids)
]
gdf_by_source = {"TDG": tdg_stops_to_plot, "C2C_TDG": c2c_tdg_stops_to_plot}
```

```python
activity_gdf["Category"] = "Unreachable"
activity_gdf.loc[~activity_gdf["Id wp"].isin(department_activity_df["Id wp"]), "Category"] = (
    "Outside departement"
)
activity_gdf.loc[activity_gdf["Id wp"].isin(distances_df["Id wp"].unique()), "Category"] = (
    "Reachable"
)
activity_gdf.groupby("Category").count()["Id wp"]
color_mapping = {"Outside departement": "gray", "Reachable": "green", "Unreachable": "purple"}
activity_gdf["color"] = activity_gdf["Category"].apply(lambda c: color_mapping[c])
```

```python
reachable_nb = sum(activity_gdf["Category"] == "Reachable")
unreachable_nb = sum(activity_gdf["Category"] == "Unreachable")
outside_nb = sum(activity_gdf["Category"] == "Outside departement")
title_html = f"""
    <h4 style="position: fixed; top: 10px; left: 50px; z-index: 9999; background: rgba(255,255,255,0.8); padding: 6px 10px; margin: 0;">
    Points d'accès d'activité CampToCamp.<br>
    - {reachable_nb} points à moins de {distance_threshold_km}km d'un arrêt de bus (<span style=\"color: green\">vert</span>)<br>
    - {unreachable_nb} points à plus de {distance_threshold_km}km d'un arrêt de bus (<span style=\"color: purple\">violet</span>)<br>
    - {outside_nb} points en dehors du département (<span style=\"color: gray\">gris</span>)
    </h4>
"""
```

```python
# Add a base OSM map centered on Grenoble
m = folium.Map(
    location=[45.1885, 5.7245], zoom_start=9, tiles="OpenStreetMap", control_scale=True
)

# Add Waymarked Trails hiking layers
folium.TileLayer(tiles="WaymarkedTrails.hiking", control=False).add_to(m)

# Add department polygon
isere_latlon_coords = [
    (lat, lon)
    for lon, lat in area_gdf[area_gdf["code"] == department_code]["geometry"]
    .values[0]
    .exterior.coords
]
folium.Polygon(
    locations=isere_latlon_coords,
    color="black",
    weight=3,
    fill_color="black",
    fill_opacity=0.25,
    fill=True,
).add_to(m)

# Add bus markers
for source in ["TDG", "C2C_TDG"]:
    fg = folium.FeatureGroup(name=f"Bus stops from {source} ({bus_stop_nb[source]})", show=True)
    fields = fields_by_source[source]
    aliases = aliases_by_source[source]
    folium.GeoJson(
        gdf_by_source[source][fields + ["geometry"]],
        zoom_on_click=True,
        marker=folium.Marker(
            icon=folium.Icon(icon="bus", prefix="fa", color=color_by_source[source])
        ),
        tooltip=folium.GeoJsonTooltip(fields=fields, aliases=aliases),
        popup=folium.GeoJsonPopup(fields=fields, aliases=aliases),
    ).add_to(fg)
    fg.add_to(m)

# Add activity markers
columns_to_display = ["Id wp", "Name wp", "nbr_topo"]
fg = folium.FeatureGroup(name="C2C activity access point", show=True)
aliases = [f"<b>Point d'accès d'activité</b><br>ID :", "Nom :", "Nombre de topos :"]
folium.GeoJson(
    activity_gdf[activity_gdf["Category"] != "Outside departement"][
        columns_to_display + ["color", "geometry"]
    ],
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(icon="star")),
    tooltip=folium.GeoJsonTooltip(fields=columns_to_display, aliases=aliases),
    popup=folium.GeoJsonPopup(fields=columns_to_display, aliases=aliases),
    style_function=lambda x: {
        "markerColor": x["properties"]["color"],
    },
).add_to(fg)
fg.add_to(m)

# Add layer control to filter layers
folium.LayerControl(collapsed=False).add_to(m)

# Add title
m.get_root().html.add_child(folium.Element(title_html))

m
```

```python
m.save(
    f"compare_c2c_tdg_bus_stops_at_distance_{distance_threshold_km}_km_department_{department_code}.html"
)
```
